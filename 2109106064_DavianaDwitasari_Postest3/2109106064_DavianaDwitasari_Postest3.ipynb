```python
data = df.work_type.value_counts()
plt.figure(figsize = (10, 8))
plt.pie(data, labels=data.index,colors=['green', 'blue', 'yellow','red','BurlyWood'], autopct="%1.1f%%")
plt.title('Grafik Pie Chart dari Tipe Perkerjaan Pasien')
plt.show()
```

### Kesimpulan:

Kesimpulan yang dapat diambil dari grafik diatas adalah tipe pekerjaan para pasien mayoritas adalah private, disusul dengan Self Employed sebanyak 16.0%, lalu Children sebanyak 13.4%, Govt_Job sebanyak 12.9%, dan yang terakhir adalah Never_worked 0.4%

```python
cor_matrix = df.corr()
sns.heatmap(cor_matrix, cmap='flare', annot=True, center=0)
plt.show()
```

```python
data = df.groupby('smoking_status').mean()
data
```

```python
x = data.index
y = data.age
plt.bar(x, y, width = 1, color=['IndianRed','HotPink', 'yellow','BurlyWood'])
plt.title('Hubungan rata rata Usia dengan Smooking Status')
plt.xlabel('Smooking Status')
plt.ylabel('Usia')
plt.xticks(rotation=90, horizontalalignment='center', fontsize=12)
plt.show()
```

### Kesimpulan

Pada diagram diatas menunjukkan bagaimana hubungan antara smooking status dengan rata rata usia dapat dilihat bahwa  kategori unknwon berada di rata-rata usia 30 an, lalu untuk kategori  formerly smoked berada di rata-rata usia 50 an, kategori Never Smoked memiliki rata-rata usia 40 an keatas, dan yang terakhir kategori smokes memiliki rata-rata usia diatas 40 an juga.

```python
sns.set_style('whitegrid')
plt.figure(figsize=(13,7))

dataplot = df.sort_values('avg_glucose_level')
sns.scatterplot(
    data   = dataplot,
    x      = 'work_type',
    y      = 'avg_glucose_level',
    hue    = 'age',
    color  = 'red'
    )

plt.xlabel('Tipe Pekerjaan')
plt.ylabel('Rata-rata Level Glukosa')
plt.title('Grafik Perbandingan Data Karyawan Pada Setiap Rata-rata Skor Training Berdasarkan Usia', fontsize = 15)
plt.show()
```

### Kesimpulan

Kesimpulan yang dapat diambil dari grafik di atas adalah terdapat data rata-rata usia pasien yang diterima pada tiap tipe perkerjaan pasien adalah berusia dari rentang 60-75 tahun dengan berdasarkan rata-rata Glukosa Level.
