## Memasukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya

```python
jumlahRecord = df.shape[0]
jumlahAttribute = df.shape[1]
namaAttribute = list(df.columns)
jumlahAttributeAngka = df.select_dtypes(['int64','float']).shape[1]

dictData = {
    'Jumlah Record' : jumlahRecord,
    'Jumlah Attribute' : jumlahAttribute,
    'Nama Attribute' : namaAttribute,
    'Jumlah Attribute Angka' : jumlahAttributeAngka
}
```

## Menampilkan isi dari dictionary

```python
dfNew = pd.DataFrame.from_dict(dictData, orient ='index').T

dfNew
```
